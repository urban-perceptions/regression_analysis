```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
%matplotlib inline
```

```python
df = pd.read_csv('./tweets_with_topics_sentiment.csv')
df
```

```python
df['dominant_topic'] = df['dominant_topic'].replace('exploring', 'explore')
```

```python
df['dominant_topic'].unique()
```

```python
df.columns
```

```python
city_dict = {'Fort Worth': 1, 'San Antonio': 1, 'Phoenix': 1, 'Las Vegas': 1,
       'Jacksonville': 1, 'San Francisco': 0, 'New York': 0, 'Washington DC': 0,
       'Boston': 0, 'Chicago': 0, 'Tampa': 1, 'Kansas City': 1, 'Raleigh': 1,
       'Philadelphia': 0, 'Pittsburgh': 0, 'New Orleans': 0, 'Los Angeles': 0}
```

```python
df['city'].unique()
```

```python
def get_xy(given_topic):
    """
    For tweets with a given topic, get lists of corresponding sentiments
    whether the tweet is inside or outside the city, and whether the tweet
    is about a shrinking or growing city.
    """
    x = []
    y = []
    for index, row in df.iterrows():
        topic = row['dominant_topic']
        sentiment = row['sentiment_numeric']
        inside = row['inside']
        city = row['city']
        growing = city_dict[city]

        if topic == given_topic and sentiment != 1:
            y.append(growing)
            x.append([sentiment, inside])
    
    return x, y
```

```python
topics = ['?', 'social', 'transportation', 'music', 'everyday life',
          'explore', 'sports', 'businesses/services', 'fun']

for topic in topics:
    x, y = get_xy(topic)
    clf = LogisticRegression(random_state=0).fit(x, y)
    score = clf.score(x, y)
    print('For topic "' + topic + '" the score is ' + str(score))
```
